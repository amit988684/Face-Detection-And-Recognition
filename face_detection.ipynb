```python
import cv2
import matplotlib.pyplot as plt
import time
%matplotlib inline
```

```python
def convertToRGB(img):
    return cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
test1 = cv2.imread("screen-shot-2014-10-09-at-3-48-23-pm.png")
gray_img = cv2.cvtColor(test1,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_img,cmap='gray')
```

```python
haar_face_cascade = cv2.CascadeClassifier('open_cv/opencv-master/data/haarcascades/haarcascade_frontalface_alt.xml')
```

```python
faces = haar_face_cascade.detectMultiScale(gray_img,scaleFactor=1.1,minNeighbors=5)
```

```python
print(faces,"Num of Faces",len(faces))
```

```python
for (x,y,w,h) in faces:
    cv2.rectangle(test1,(x,y),(x+w,y+h),(0,255,0),10)
```

```python
plt.imshow(convertToRGB(test1))
```

```python


# Helper Function
def detect_faces(colored_img,cascade_c,scale_f=1.1):
    img_copy = colored_img.copy()
    gray_img = cv2.cvtColor(img_copy,cv2.COLOR_BGR2GRAY)
    faces = cascade_c.detectMultiScale(gray_img,scaleFactor=scale_f,minNeighbors=5);
#     print(faces,"Num of Faces",len(faces))
    
    for (x,y,w,h) in faces:
        cv2.rectangle(img_copy,(x,y),(x+w,y+h),(0,255,0),6)
        
    return img_copy
```

```python
imgg = cv2.imread("860_main_beauty.png")
# haar_cat_cascade = cv2.CascadeClassifier("open_cv/opencv-master/data/haarcascades/haarcascade_frontalcatface.xml")
# haar = cv2.CascadeClassifier("open_cv/opencv-master/data/haarcascades/haarcascade_righteye_2splits.xml")
faces_detect_img = detect_faces(imgg,haar_face_cascade)
```

```python
plt.imshow(convertToRGB(faces_detect_img))
```
